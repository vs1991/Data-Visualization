```python
from scipy.stats import shapiro
from scipy import stats
```

```python
def univar_analys(a):
    
    print('Mean value:',a.mean())
    print('Median value:',a.median())
    print('Standard deviation:',a.std())
    print('Mode value:',a.median())
    print('Q1:',a.quantile(0.25))
    print('Q3:',a.quantile(0.75))
    print('Minimum:',a.min())
    print('Maximum:',a.max())
    print('Skewness:',a.skew())
    print('Shapriro test:',shapiro(a))
    print('Normal test:',stats.normaltest(a))
    print('Kurtosis:',a.kurtosis())
    
    if a.skew()>1.5:
        print('Positively skewed')
    else:
        print('Negatively skewed')
    
    plt.figure(figsize=(20,5))
    sns.distplot(a,bins=int((96453)**0.33)*2)
    plt.axvline(a.mean(), color='r', linestyle='--', linewidth=1)
    plt.axvline(a.median(), color='black', linestyle='--', linewidth=1)
    plt.axvline(a.mode().get_values()[0], color='g', linestyle='--', linewidth=1)

    
    return 
```

```python

# generate gaussian data
from numpy.random import seed
from numpy.random import randn
from numpy import mean
from numpy import std
# seed the random number generator
seed(1)
# generate univariate observations
data = 5 * randn(100) + 50
# summarize
print('mean=%.3f stdv=%.3f' % (mean(data), std(data)))
```

```python
seed(1)
# generate univariate observations
data = 5 * randn(100) + 50
# histogram plot
plt.hist(data)
plt.show()
```

```python
from statsmodels.graphics.gofplots import qqplot
data = 5 * randn(100) + 50
# q-q plot
qqplot(data, line='s')
plt.show()
```

```python
stat, p = shapiro(data)
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
	print('Sample looks Gaussian (fail to reject H0)')
else:
	print('Sample does not look Gaussian (reject H0)')
```
